```python
data_AQI = data[['State', 'Date', 'NO2 AQI', 'O3 AQI', 'SO2 AQI', 'CO AQI']]
data_AQI_State = data_AQI.groupby('State').mean()
data_AQI_State.reset_index(inplace=True)

plt.rcParams["figure.figsize"] = (18, 12) 

x = np.arange(1,11)

fig, axs = plt.subplots(4, 1)

# Plot for NO2
data_AQI_State.sort_values(by = 'NO2 AQI', ascending = False, inplace = True)
barplot1 = axs[0].bar(x, 'NO2 AQI', data=data_AQI_State[:10], label = 'NO2')
xlabels = data_AQI_State.loc[:10, 'State']
plt.sca(axs[0])
#plt.xticks(x,xlabels) # x axis marks

# Plot for O3
data_AQI_State.sort_values(by = 'O3 AQI', ascending = False, inplace = True)
barplot2 = axs[1].bar(x, 'O3 AQI', data=data_AQI_State[:10], label = 'O3', color = 'red')
xlabels = data_AQI_State.loc[:10, 'State']
plt.sca(axs[1])
#plt.xticks(x, xlabels) # x axis marks

# Plot for SO2
data_AQI_State.sort_values(by = 'SO2 AQI', ascending = False, inplace = True)
barplot3 = axs[2].bar(x, 'SO2 AQI', data=data_AQI_State[:10], label = 'SO2', color = 'green')
xlabels = data_AQI_State.loc[:10, 'State']
plt.sca(axs[2])
#plt.xticks(x, xlabels) # x axis marks

# Plot for CO
data_AQI_State.sort_values(by = 'CO AQI', ascending = False, inplace = True)
barplot4 = axs[3].bar(x, 'CO AQI', data=data_AQI_State[:10], label = 'CO', color = 'purple')
xlabels = data_AQI_State.loc[:10, 'State']
plt.sca(axs[3])
#plt.xticks(x, xlabels) # x axis marks

axs[0].tick_params(axis='both', which='major', labelsize=12) 
axs[1].tick_params(axis='both', which='major', labelsize=12)
axs[2].tick_params(axis='both', which='major', labelsize=12)
axs[3].tick_params(axis='both', which='major', labelsize=12)


# Adding value for each bar
i = 0
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    global i
    for rect in rects:
        height = rect.get_height()
        axs[i].annotate('{:.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', size = 12)
    i = i + 1

autolabel(barplot1)
autolabel(barplot2)
autolabel(barplot3)
autolabel(barplot4)

# Space between charts
plt.subplots_adjust(hspace=0.3) 

# Adding plot titles and axis titles
axs[0].set_title('TOP 10 biggest polluters by mean value of AQI for Nitrogen Dioxide (NO2) for 2000-2021', fontsize = 14)
axs[1].set_title('TOP 10 biggest polluters by mean value of AQI for Ozone (O3) for 2000-2021', fontsize = 14)
axs[2].set_title('TOP 10 biggest polluters by mean value of AQI for Sulfur Dioxide (SO2) for 2000-2021', fontsize = 14)
axs[3].set_title('TOP 10 biggest polluters by mean value of AQI for Carbon Monoxide (CO) for 2000-2021', fontsize = 14)
fig.text(0.1, 0.5, 'Mean value of AQI', ha='center', va='center', rotation='vertical', fontsize = 14)
```

```python
data.columns
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=data['CO AQI'], orient='h')
plt.show()
```

```python
Q1=np.percentile(data['CO AQI'], 25, interpolation='midpoint')
Q2=np.percentile(data['CO AQI'], 50, interpolation='midpoint')
Q3=np.percentile(data['CO AQI'], 75, interpolation='midpoint')
IQR = Q3-Q1
upper_limit=round(Q3+(1.5 * IQR), 4)
lower_limit=round(Q1-(1.5 * IQR), 4)
outlier=[]
for x in data['CO AQI']:
    if (x<lower_limit) or (x>upper_limit):
        outlier.append(x)
print(' Q1: ',Q1,'\n',
      'Q2: ',Q2,'\n',
      'Q3: ',Q3)
print('Upper limit: ',upper_limit)
print('Lower limit: ',lower_limit)
print('Number of outliers in CO AQI: ',len(outlier))
```

```python
data['CO AQI']=np.where(data['CO AQI']>upper_limit,upper_limit,np.where(data['CO AQI']<lower_limit,lower_limit,data['CO AQI']))
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=data['CO AQI'], orient='h')
plt.show()
```
