```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
auto=pd.read_csv('tele.csv')
```

```python
auto
```

# compare churn count with respect to gender

```python
sns.countplot(x="gender",hue="Churn",data=auto)
```

# find out how many female senior citizens there in the dataset

```python
abc=auto.loc[(auto['gender']=='Female') & (auto['SeniorCitizen']==1)]
abc
sns.countplot(abc['SeniorCitizen'])
plt.xticks
plt.grid()
```

# compare tenure with total charges

```python
plt.scatter(auto["tenure"],auto["TotalCharges"],s=10,marker='^')
plt.xlabel("tenure")
plt.ylabel("TotalCharges")
plt.title("GRAPH BETWEEN TENURE VS TOTAL CHARGES")
```

# findout which contract is preferred by the senior citizen

```python
ab=auto.loc[(auto['SeniorCitizen']==1)]
ab.head()
sns.countplot(ab["SeniorCitizen"],hue=ab["Contract"])
plt.xticks(visible= False)
plt.grid()
```
