```python
auto.set_PassengerId(auto['passa'], inplace=True)
auto
```

```python
auto=auto.drop['PassengerId']
```

```python
auto.isnull().sum()
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
auto['Age'] = auto[['Age','Pclass']].apply(impute_age,axis=1)
```
